```python
import cv2 
from google.colab.patches import cv2_imshow
```

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        
        gray_image = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        gray_scale_image = 255 - gray_image
        
        down_width = 32
        down_height = 32
        down_points = (down_width, down_height)
        resized_down = cv2.resize(gray_scale_image, down_points, 
                                  interpolation= cv2.INTER_LINEAR)
        norm = np.zeros((100,100))
        img = cv2.normalize(gray_scale_image,  norm, 0, 255, cv2.NORM_MINMAX)
        

        if img is not None:
            images.append(img)
    return images
```

```python
for i in range(10):
  images = load_images_from_folder(f'/content/drive/MyDrive/Colab Notebooks/spectrograms/{i}') 
  images.append(i)
  
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.iloc[:,3:5].values
y = data.iloc[:,-1].values
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_main, y_main, test_size=1000, stratify=y_main)
X_train.shape,X_val.shape
```
